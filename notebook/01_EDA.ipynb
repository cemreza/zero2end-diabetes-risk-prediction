```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (8, 5)

df = pd.read_csv("../data/raw/diabetes_012_health_indicators_BRFSS2015.csv")

df.head() 
```

```python
df.shape
```

```python
df.info()
```

```python
df['Diabetes_012'].value_counts(normalize=True) * 100
```

```python
df['Diabetes_012'].value_counts().plot(kind="bar")
plt.title("Diabetes_012 class distribution")
plt.xlabel("Class (0=No Diabetes, 1=Prediabetes, 2=Diabetes)")
plt.ylabel("Count")
plt.show()
```

```python
df['Diabetes_binary'] = df['Diabetes_012'].apply(lambda x: 0 if x == 0 else 1)

df['Diabetes_binary'].value_counts(normalize=True) * 100
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(), cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.hist(df['BMI'], bins=40)
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df['Age'], bins=40)
plt.title("Age Distribution")
plt.xlabel("Age Group (coded 1â€“13)")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x='Diabetes_binary', y='BMI', data=df)
plt.title("BMI vs Diabetes Risk")
plt.show()
```

```python
sns.boxplot(x='Diabetes_binary', y='GenHlth', data=df)
plt.title("General Health Score vs Diabetes Risk")
plt.show()
```
